# 4.sep parameter

##### generally in read_csv the sep parameter = ',' but when we use like tab seperated values file then to differentiate between column  we use this sep parameter as shown below

```python
pd.read_csv('movie_titles_metadata.tsv')
```

##### we are not able to understand what is written above b/c all columns have been mixed up b/c there were no comma in the file from where read_csv can separate on basis of comma so we will use sep parameter as here the values are separated by tab 

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t')
```

#### in the above o/p we can see that one of the rows only have been named as column name which is wrong so we will use names parameter to correct it there in names parameter we will pass a list of columns

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sr_no','name','release_year','ratings','votes','genres'])
```

# 13. converters

```python
pd.read_csv('IPL Matches 2008-2020.csv')
```

### ab jahan bhi royal challengers bengaluru hai wahan pe i want RCB then here  convertors come into roles

```python
def rename(name):
    if name=="Royal Challengers Bangalore":
        return "RCB"
    elif name=="Chennai Super Kings":
        return "CSK"
    elif name=="Kolkata Knight Riders":
        return "KKR"
    else:
        return name
```

```python
rename("Kolkata Knight Riders")
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename,'team2':rename,'toss_winner':rename,'winner':rename})
```
